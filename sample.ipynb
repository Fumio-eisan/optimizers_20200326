```python
#%%
import numpy as np
import optimizers
import matplotlib.pyplot as plt
```

```python
max_iterations = 100

y7=[]
y8=[]
y9=[]

v7=[]
v8=[]
v9=[]
optimizer = optimizers.SGD()

x = 100
optimizer.lr =0.1
for i in range(max_iterations):
    x = optimizer.update(x)
    y7.append(x)
    v7.append(optimizer.lr)
    
x = 100
optimizer.lr =0.01
for i in range(max_iterations):
    x = optimizer.update(x)
    y8.append(x)
    v8.append(optimizer.lr)

x = 100
optimizer.lr =0.9
for i in range(max_iterations):
    x = optimizer.update(x)
    y9.append(x)
    v9.append(optimizer.lr)
```

```python
# %%

x = np.arange(max_iterations)
plt.plot(x, y7, label='lr=0.1')
plt.plot(x, y8, label='lr=0.01')
plt.plot(x, y9, label='lr=0.9')
plt.xlabel("iterations")
plt.ylabel("y")
plt.ylim(-5, 120)
plt.legend()
plt.show()
```

```python
max_iterations = 100
x = 100
y1=[]
y2=[]
y3=[]

v1=[]
v2=[]
v3=[]
#optimizers = optimizers.SGD()
optimizer = optimizers.Momentum()

optimizer.alpha =0.8
for i in range(max_iterations):
    x = optimizer.update(x)
    y1.append(x)
    v1.append(optimizer.v)
    
x=100
optimizer.alpha =0.1
for i in range(max_iterations):
    x = optimizer.update(x)
    y2.append(x)
    v2.append(optimizer.v)    
    
x=100
optimizer.alpha =1
for i in range(max_iterations):
    x = optimizer.update(x)
    y3.append(x)
    v3.append(optimizer.v)        
```

```python
# %%
import matplotlib.pyplot as plt
x = np.arange(max_iterations)
plt.plot(x, y1, label='alpha=0.8')
plt.plot(x, y2, label='alpha=0.1')
plt.plot(x, y3, label='alpha=1')
plt.xlabel("iterations")
plt.ylabel("y")
plt.ylim(-5, 120)
plt.legend()
plt.show()
```

```python
plt.plot(x, v1, label='alpha=0.8')
plt.plot(x, v2, label='alpha=0.1')
plt.plot(x, v3, label='alpha=1')
plt.xlabel("iterations")
plt.ylabel("v")
plt.ylim(-10, 10)
plt.legend()
plt.show()
```

```python
#AdaGrad
max_iterations = 100
x = 100
y4=[]
y5=[]
y6=[]

v4=[]
v5=[]
v6=[]

optimizer = optimizers.AdaGrad()

optimizer.h0 =10
for i in range(max_iterations):
    x = optimizer.update(x,)
    y4.append(x)
    v4.append(optimizer.v)     
    
x=100
optimizer.h0 =100
for i in range(max_iterations):
    x = optimizer.update(x,)
    y5.append(x)
    v5.append(optimizer.v)  

x=100
optimizer.h0 =1000
for i in range(max_iterations):
    x = optimizer.update(x,)
    y6.append(x)
    v6.append(optimizer.v)      
```

```python
# %%
import matplotlib.pyplot as plt
x = np.arange(max_iterations)
plt.plot(x, y4, label='eta_0=10')
plt.plot(x, y5, label='eta_0=100')
plt.plot(x, y6, label='eta_0=1000')
plt.xlabel("iterations")
plt.ylabel("y")
plt.ylim(-5, 120)
plt.legend()
plt.show()
```

```python
plt.plot(x, v4, label='h0=10')
plt.plot(x, v5, label='h0=100')
plt.plot(x, v6, label='h0=1000')
plt.xlabel("iterations")
plt.ylabel("v")
plt.legend()
plt.show()
```
